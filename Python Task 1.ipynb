```python
# Write a program to check whether a string is a palindrome or not using a function.
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    for i in range(len(s) // 2):
        if s[i] != s[-(i + 1)]:
            return False
    return True

text = input("Enter string: ")
if is_palindrome(text):
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
#20. Use the map() function to square all numbers in a list [1, 2, 3, 4, 5].
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

```python
#Use the filter() function to keep only even numbers from [1, 2, 3, 4, 5, 6].
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)
```

```python
#Use the reduce() function (from functools) to calculate the product of all numbers in [2, 3,4, 5].
from functools import reduce
numbers = [2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

```python
#Write a function that accepts a dictionary of student names and marks, and returns the name of the student with the highest marks.
def top_student(marks_dict):
    return max(marks_dict, key=marks_dict.get)
students = {"Alice": 85, "Bob": 92, "Charlie": 78}
print(top_student(students))
```

```python
#Create a list of tuples containing student names and marks. Sort it by marks using a lambda function.
students = [("Alice", 85), ("Bob", 92), ("Charlie", 78)]
students.sort(key=lambda x: x[1])
print(students)
```

```python
#25. Write a Python program that combines all the above:
#● Takes a list of numbers
#● Filters only even numbers
#● Squares them using map()
#● Finds the total using reduce()

from functools import reduce
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
squared = list(map(lambda x: x**2, evens))
total = reduce(lambda x, y: x + y, squared)
print(total)
```
