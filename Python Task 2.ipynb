```python
#Write a simple calculator using if-elif-else for +, -, ×, ÷.
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    print("Result:", num1 + num2)

elif operator == "-":
    print("Result:", num1 - num2)

elif operator == "*":
    print("Result:", num1 * num2)

elif operator == "/":
    if num2 != 0: #num2 != 0 means “num2 is not equal to zero
        print("Result:", num1 / num2)
    else:
        print("Error: Division by zero")

else:
    print("Invalid operator")
```

```python
#Write a function to calculate the factorial of a number.
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
print(factorial(5))
```

```python
#Write a function that returns the square of all numbers in a list.


def square_list(numbers):
    return [num ** 2 for num in numbers]

nums = [1, 2, 3, 4, 5]
print("Squared numbers:", square_list(nums))
```

```python
#Write a function that accepts any number of arguments and prints their sum.
def sum_numbers(*args):
    total = 0
    for num in args:
        total += num
    print("Sum:", total)

# Example usage
sum_numbers(2, 5, 7)        # Output: Sum: 14
sum_numbers(1, 2, 3, 4, 5)  # Output: Sum: 15
```

```python
#Write a function that takes a dictionary of marks and returns the topper’s name.
def get_topper(marks):
    # Returns the key (student) with the maximum value (mark)
    return max(marks, key=marks.get)

# Example usage
students = {"Faiz": 85, "Barsha": 92, "Nandita": 78}
topper = get_topper(students)
print("Topper:", topper)
```
